```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sas
data = pd.read_csv('Housing.csv')
data.head()
```

```python
sas.pairplot(data)
```

```python
corr = data.corr()
sas.heatmap(corr, annot= True)
```

```python
data.plot(kind= 'scatter', x = 'area', y = 'price', figsize = (6 , 6), color = 'red')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['area']), y= data['price'])
print(model.intercept_)
print(model.coef_)
```

```python
model.score(X = pd.DataFrame(data['area']), y= data['price'])
```

```python
data.plot(kind= 'scatter', x = 'area', y= 'price', figsize = (5,5), color = 'red')
y_pred = model.predict(X = pd.DataFrame(data['area']))
plt.plot(data['area'], y_pred, color = 'blue')
```

```python
#Mo hinh bac 2
model = linear_model.LinearRegression()
X = pd.DataFrame([data['area'], data['area']**2]).T
model.fit(X, y= data['price'])
print('Intercept cua mo hinh:', model.intercept_)
print('Coefficient cua mo hinh', model.coef_)
print('Do chinh xac cua mo hinh', model.score(X, y = data['price']))

#Ve do thi mo hinh bac 2
import numpy as np
range = np.arange(2000,16000,10)
data.plot(kind= 'scatter', x = 'area', y= 'price', figsize = (5,5), color = 'green')
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'red')
```
